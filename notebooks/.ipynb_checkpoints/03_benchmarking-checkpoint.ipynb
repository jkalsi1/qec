# Benchmarking: Shor vs Surface Code

```python
from qec.benchmark import compare_codes
from qec.visualize import plot_threshold
```

```python
error_rates = [0.001, 0.005, 0.01, 0.02, 0.05, 0.1]
df = compare_codes(error_rates, error_type="depolarizing", shots=512)
print(df.to_string(index=False))
```

```python
fig = plot_threshold(df, "Shor vs Surface Code Threshold")
display(fig)
```

## Analysis
Below the error threshold, both codes reduce the logical error rate compared to an unprotected qubit. The surface code generally outperforms the Shor code at low error rates due to its higher distance-to-qubit-count ratio and more efficient MWPM decoding.

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qec.shor import encode_shor, measure_syndrome_shor
from qec.noise import make_noise_model

def run_shor_no_decode_circuit(p, shots=512):
    noise_model = make_noise_model("bit_flip", p) if p > 0 else None
    simulator = AerSimulator(noise_model=noise_model) if noise_model else AerSimulator()

    q   = QuantumRegister(9, 'q')
    anc = QuantumRegister(6, 'anc')
    syn = ClassicalRegister(6, 'syn')
    dat = ClassicalRegister(9, 'dat')
    qc  = QuantumCircuit(q, anc, syn, dat)

    encode_shor(qc, q)
    measure_syndrome_shor(qc, q, anc, syn)

    # Apply correction
    with qc.if_test((syn, 1)):  qc.x(q[0])
    with qc.if_test((syn, 3)):  qc.x(q[1])
    with qc.if_test((syn, 2)):  qc.x(q[2])
    with qc.if_test((syn, 4)):  qc.x(q[3])
    with qc.if_test((syn, 12)): qc.x(q[4])
    with qc.if_test((syn, 8)):  qc.x(q[5])
    with qc.if_test((syn, 16)): qc.x(q[6])
    with qc.if_test((syn, 48)): qc.x(q[7])
    with qc.if_test((syn, 32)): qc.x(q[8])

    # Measure all 9 data qubits — no decode circuit
    for i in range(9):
        qc.measure(q[i], dat[i])

    tqc = transpile(qc, simulator)
    counts = simulator.run(tqc, shots=shots).result().get_counts()

    errors = 0
    for bitstring, count in counts.items():
        # Format: "dat[8]..dat[0] syn[5]..syn[0]"
        dat_str = bitstring.split(' ')[0]
        # dat[i] = dat_str[8-i]
        dat_list = [int(dat_str[8 - i]) for i in range(9)]

        # Each block: majority vote to get block value
        b0 = 1 if (dat_list[0] + dat_list[1] + dat_list[2]) >= 2 else 0
        b1 = 1 if (dat_list[3] + dat_list[4] + dat_list[5]) >= 2 else 0
        b2 = 1 if (dat_list[6] + dat_list[7] + dat_list[8]) >= 2 else 0

        # Logical bit: majority of block values
        # Valid |0> codeword has all blocks in same state after correction
        # But since blocks are independent in X basis, we need parity check
        # XOR of all block values: 0 means even number of 1-blocks = logical 0
        # This works because each block independently encodes the same logical bit
        logical_bit = 1 if (b0 + b1 + b2) >= 2 else 0
        if logical_bit == 1:
            errors += count

    return errors / shots

print("p=0.0  no decode circuit:", run_shor_no_decode_circuit(0.0, shots=256))
print("p=0.001 no decode circuit:", run_shor_no_decode_circuit(0.001, shots=512))
print("p=0.01  no decode circuit:", run_shor_no_decode_circuit(0.01, shots=512))
```
